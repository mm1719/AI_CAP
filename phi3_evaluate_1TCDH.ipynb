```python
import sys

# Helper function to execute a list of code strings
def execute_code_blocks(task_id, code_blocks, context):
    flag = True
    for code in code_blocks:
        try:
            exec(code, context)
        except Exception as e:
            #print(f"\nException task ID: {task_id}, Exception: {e}")
            flag = False
            break

    return flag
```

```python
# @title 預設標題文字
# Test with all code
from tqdm import tqdm

data_path = '/content/drive/My Drive/Colab Notebooks/AI_CAP/data/'
eval_path= '/content/drive/My Drive/Colab Notebooks/AI_CAP/eval/'
with open(os.path.join(data_path, f'results_origin.json'), 'r') as f:
    solutions_data = json.load(f)

passed_cnt = 0
failed_cnt = 0
exception_cnt = 0

# Open a log file for writing
with open(os.path.join( eval_path, f"Phi3_base_eval.log"), "w") as log_file:
    print("test")
    for i in tqdm(range(len(solutions_data)), desc="Processing code"):
        code = solutions_data[i]
        code_blocks = [code[2]][0]
        test_list = code[1]
        task_id = i + 11

        if task_id in exception_list:
            continue

        log_file.write(f"Task ID: {task_id}\n")
        log_file.write(f"Generated Code: {code_blocks}\n")

        context = {}
        flag = execute_code_blocks(task_id, code_blocks, context)

        if not flag:
            exception_cnt += 3
            continue

        for test in test_list:
            try:
                exec(test, context)
                log_file.write(f"Test passed: {test}\n")
                passed_cnt += 1
                #print(f"Test passed: {test}")
            except AssertionError:
                log_file.write(f"Test failed: {test}\n")
                #print(f"Test failed: {test}")
                failed_cnt += 1
            except Exception as e:
                log_file.write(f"Test raised an exception: {test}\nException: {e}\n")
                #print(f"\nException task ID: {task_id}")
                #print(f"Test raised an exception: {test}\nException: {e}")
                exception_cnt += 1

    #print("-" * 80)
```

```python
# Success 1, exception 2
def reverse_vowels(s):
	vowels = ["a", "e", "i", "o", "u", "A", "E", "I", "O", "U"]
	q = 0
	h = len(s) - 1
	s = list(s)
	while q < h:
 		if s[q] in vowels and s[h] in vowels:
 			s[q], s[h] = s[h], s[q]
 			q += 1
 			h -= 1
	return "".join(s)

assert reverse_vowels("ab") == "ba"
```

```python
# exception 3
def shell_sort(arr):
	for (i, el) in enumerate(arr):
		position = i + 1
		while position < len(arr):
			key = arr[position]
			j = position - 1
			while j >= 0 and key < arr[j]:
				arr[j + 1] = arr[j]
				j -= 1
			arr[j + 1] = key
		position += 1
	return arr

assert shell_sort([32, 30, 16, 96, 82, 83, 74]) == [16, 30, 32, 74, 82, 83, 96]
```

```python
# exception 3
def same_order(list1, list2):
	index1 = 0
	index2 = 0
	while index1 < len(list1) and index2 < len(list2):
		if list1[index1] == list2[index2]:
			index1 += 1
			index2 += 1
		else:
			index1 = index1
	return index1 == len(list1)

assert same_order(["red","green","black","orange"],["red","pink","green","white","black"])==True
```
