# Use llama3 to solve coding problem

````python
# Solve Problem Function
def solve_code_problem(task_id, description, test_list, test_setup_code):
    prompt = f"Task ID: {task_id}\nDescription: {description}\nTest List:\n{test_list}\nTest Setup Code:{test_setup_code}"

    result = client.chat.completions.create(
        messages=[
            {
                'role': 'user',
                'content': prompt
            }
        ],
        model="llama3-70b-8192"
    )
    #Markdown(result.choices[0].message.content)
    #print(result.choices[0].message.content)
    #print(type(result.choices[0].message.content))
    return result.choices[0].message.content
````

````python
def extract_code_block(text):
    import re
    # Regex to capture code blocks starting with "```python" or just "python\n"
    code_blocks = re.findall(r'```python\n(.*?)\n```|python\n(.*?)\n```|```\npython\n(.*?)\n```|```\n(.*?)\n```', text, re.IGNORECASE | re.DOTALL)
    # Flatten the list of tuples and filter out empty strings
    code_blocks = [code for block in code_blocks for code in block if code]

    # Remove assert statements from code blocks, and remove the empty block
    cleaned_code_blocks = []
    for block in code_blocks:
        lines = block.split('\n')
        inside_assert_block = False
        cleaned_lines = []
        for line in lines:
            stripped_line = line.strip()
            if stripped_line.startswith('assert'):
                inside_assert_block = True
            if not inside_assert_block:
                cleaned_lines.append(line)
            if inside_assert_block and stripped_line == '':
                inside_assert_block = False
        cleaned_block = '\n'.join(cleaned_lines).strip()
        if cleaned_block:
            cleaned_code_blocks.append(cleaned_block)

    return cleaned_code_blocks

    return cleaned_code_blocks

block = extract_code_block(solutions_data[6]['solution'])
print(block)
print(block[0])
````

# Run Solution Code

````python
import sys

# Helper function to execute a list of code strings
def execute_code_blocks(task_id, code_blocks, context):
    flag = True
    for code in code_blocks:
        try:
            exec(code, context)
        except Exception as e:
            #print(f"\nException task ID: {task_id}, Exception: {e}")
            flag = False
            break

    return flag
````

````python
# Go through all test
import math, re
from tqdm import tqdm

code_list = [100, 200, 300, 400, 500]
code_path = '/content/drive/My Drive/Colab Notebooks/AI_CAP/data/code/'

failed_cnt = 0
exception_cnt = 0

for i in tqdm(code_list, desc="Processing code files"):
    with open(os.path.join(code_path, f'code_{i}.json'), 'r') as f:
        code_data = json.load(f)

    for j in tqdm(range(len(code_data)), desc=f"Processing tasks in code_{i}.json"):
        code = code_data[j]
        task_id = code['task_id']
        code_blocks = code['code']
        test_list = code['test_list']
        test_setup_code = code['test_setup_code']

        if task_id in more_than_one_block:
            print(f"\nTask {task_id} has more than one code block. Skipping.")
            continue

        if task_id in time_out_block:
            print(f"\nTask {task_id} has time out code block. Skipping.")
            continue

        context = {}
        flag = execute_code_blocks(task_id, code_blocks, context)

        if not flag:
            exception_cnt += 3
            continue

        if test_setup_code:
            exec(test_setup_code, context)

        #print(f"Task ID: {task_id}")

        for test in test_list:
            try:
                exec(test, context)
                #print(f"Test passed: {test}")
            except AssertionError:
                #print(f"Test failed: {test}")
                failed_cnt += 1
            except Exception as e:
                #print(f"\nException task ID: {task_id}")
                #print(f"Test raised an exception: {test}\nException: {e}")
                exception_cnt += 1

        #print("-" * 80)
````
