```python
import sys

# Helper function to execute a list of code strings
def execute_code_blocks(task_id, code_blocks, context):
    flag = True
    for code in code_blocks:
        try:
            exec(code, context)
        except Exception as e:
            #print(f"\nException task ID: {task_id}, Exception: {e}")
            flag = False
            break

    return flag
```

```python
# Test with first 5 code
data_path = '/content/drive/My Drive/Colab Notebooks/AI_CAP/data/'
with open(os.path.join(data_path, f'results_with_tune.json'), 'r') as f:
    solutions_data = json.load(f)

for i in range(5):
    code = solutions_data[i]
    code_blocks = [code[1]]
    test_list = code[0]
    task_id = i + 11

    context = {}
    execute_code_blocks(task_id, code_blocks, context)

    print(f"Task ID: {task_id}")

    for test in test_list:
        try:
            exec(test, context)
            print(f"Test passed")
        except AssertionError:
            print(f"Test failed: {test}")
        except Exception as e:
            print(f"Test raised an exception: {test}\nException: {e}")

    print("-" * 80)
```

```python
# Test with all code
from tqdm import tqdm

data_path = '/content/drive/My Drive/Colab Notebooks/AI_CAP/data/'
with open(os.path.join(data_path, f'results_with_tune.json'), 'r') as f:
    solutions_data = json.load(f)

failed_cnt = 0
exception_cnt = 0

for i in tqdm(range(len(solutions_data)), desc="Processing code"):
    code = solutions_data[i]
    code_blocks = [code[1]]
    test_list = code[0]
    task_id = i + 11

    if task_id in exception_list:
        continue

    context = {}
    flag = execute_code_blocks(task_id, code_blocks, context)

    if not flag:
        exception_cnt += 3
        continue

    for test in test_list:
        try:
            exec(test, context)
            #print(f"Test passed: {test}")
        except AssertionError:
            #print(f"Test failed: {test}")
            failed_cnt += 1
        except Exception as e:
            #print(f"\nException task ID: {task_id}")
            #print(f"Test raised an exception: {test}\nException: {e}")
            exception_cnt += 1

    #print("-" * 80)
```

```python
def permute_string(str1):
    res = [str1]
    while (len(res) != len(str1)):
        for i in range(len(res)):
            for j in range(len(str1)):
                if str1[j] not in res[i]:
                    res.append(res[i] + str1[j])
    return (res)
assert permute_string('abc')==['abc', 'bac', 'bca', 'acb', 'cab', 'cba']
```

```python
def is_tree_balanced(root):
    if root is None:
        return True
    left_height = height(root.left)
    right_height = height(root.right)
    if abs(left_height - right_height) <= 1:
        return (is_tree_balanced(root.left) and
                is_tree_balanced(root.right))
    return False

assert is_tree_balanced(root) == False
```
